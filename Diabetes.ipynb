# outliers Detection and treatment

```python
plt.figure(figsize=(15,10))

plt.subplot(4,4,1)
sns.boxplot(data['Pregnancies'])

plt.subplot(4,4,2)
sns.boxplot(data['Glucose'])

plt.subplot(4,4,3)
sns.boxplot(data['BloodPressure'])

plt.subplot(4,4,4)
sns.boxplot(data['SkinThickness'])

plt.subplot(4,4,5)
sns.boxplot(data['Insulin'])

plt.subplot(4,4,6)
sns.boxplot(data['BMI'])

plt.subplot(4,4,7)
sns.boxplot(data['DiabetesPedigreeFunction'])


plt.subplot(4,4,8)
sns.boxplot(data['Age'])
```

```python
* apart from glucose all other attributes have a outliers .these lower level and upper level outliers
are replaced with 5th(.05) and 95th(.95) percentile respectively
```

```python
* numpy.clip()function is used to clip(limit)the values in an array.

SYNTAX=numpy.clip(a,a_min,a_max,out=None)
parameters,
   a :Array containing elements to clip
a_min:Minimum value
a_Max:Maximum value
if clipping is not performed on the lower interval edge.not more than one of a_min and a_max may be None  
```

```python
data['Pregnancies']=data['Pregnancies'].clip(lower=data['Pregnancies'].quantile(.05),upper=data['Pregnancies'].quantile(.95))
```

```python
data['Glucose']=data['Glucose'].clip(lower=data['Glucose'].quantile(.05),upper=data['Glucose'].quantile(.95))
```

```python
data['BloodPressure']=data['BloodPressure'].clip(lower=data['BloodPressure'].quantile(.05),upper=data['BloodPressure'].quantile(.95))
```

```python
data['SkinThickness']=data['SkinThickness'].clip(lower=data['SkinThickness'].quantile(.05),upper=data['SkinThickness'].quantile(.95))
```

```python
data['Insulin']=data['Insulin'].clip(lower=data['Insulin'].quantile(.05),upper=data['Insulin'].quantile(.95))
```

```python
data['BMI']=data['BMI'].clip(lower=data['BMI'].quantile(.05),upper=data['BMI'].quantile(.95))
```

```python
data['DiabetesPedigreeFunction']=data['DiabetesPedigreeFunction'].clip(lower=data['DiabetesPedigreeFunction'].quantile(.05),upper=data['DiabetesPedigreeFunction'].quantile(.95))
```

```python
data['Age']=data['Age'].clip(lower=data['Age'].quantile(.05),upper=data['Age'].quantile(.95))
```

```python
plt.figure(figsize=(15,10))

plt.subplot(4,4,1)
sns.boxplot(data['Pregnancies'])

plt.subplot(4,4,2)
sns.boxplot(data['Glucose'])

plt.subplot(4,4,3)
sns.boxplot(data['BloodPressure'])

plt.subplot(4,4,4)
sns.boxplot(data['SkinThickness'])

plt.subplot(4,4,5)
sns.boxplot(data['Insulin'])

plt.subplot(4,4,6)
sns.boxplot(data['BMI'])

plt.subplot(4,4,7)
sns.boxplot(data['DiabetesPedigreeFunction'])


plt.subplot(4,4,8)
sns.boxplot(data['Age'])
```

```python
* as we can still see the outliers in skinthickness and insulin .lets try manipulating the percentile values.
```

```python
data['SkinThickness']=data['SkinThickness'].clip(lower=data['SkinThickness'].quantile(.07),upper=data['SkinThickness'].quantile(.93))
```

```python
data['Insulin']=data['Insulin'].clip(lower=data['Insulin'].quantile(.21),upper=data['Insulin'].quantile(.80))
```

```python
plt.figure(figsize=(20,15))

plt.subplot(4,4,4)
sns.boxplot(data['SkinThickness'])

plt.subplot(4,4,5)
sns.boxplot(data['Insulin'])
```

```python
* still we have  a oulier for insulin so lets try some more manipulating percentile value
```

```python
data['Insulin']=data['Insulin'].clip(lower=data['Insulin'].quantile(.25),upper=data['Insulin'].quantile(.75))
```

```python
plt.figure(figsize=(20,15))

plt.subplot(4,4,5)
sns.boxplot(data['Insulin'])
```

```python
* as we can see there is no outlier for the insulin,,, we are good to go further
```
